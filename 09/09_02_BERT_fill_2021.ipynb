```
#pipeline-t fogunk használni
from transformers import pipeline
```

```
#Huggingface magyar nyelvű BERT modelljét használjuk
unmaskerHU= pipeline('fill-mask', model='SZTAKI-HLT/hubert-base-cc')
```

```
#Ez a tanításnál is használt maszkolásos feladatot teszteljük:
unmaskerHU("Az [MASK] értelme.")
```

```
#Ugyanazt megnézzük angol model esetén is.
unmasker = pipeline('fill-mask', model='bert-base-uncased')
```

```
unmasker("Hello I'm a [MASK] model.")
```

```
unmasker("The man worked as a [MASK].")
```

```
unmasker("The woman worked as a [MASK].")
```
